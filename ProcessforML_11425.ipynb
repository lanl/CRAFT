```python
df=pd.DataFrame()
for i in list(range(0,2000)):
    print(i)
    my_file=DRIVE1+"BCI_ml_"+str(i)+"h1.csv"
    if os.path.isfile(my_file):
        one=pd.read_csv(my_file)    
        one['Date']=pd.DatetimeIndex(one["Date"])
        print(one["Date"])
        one['Day']=one["Date"].dt.day
        one['Month']=one["Date"].dt.month
        one['Year']=one["Date"].dt.year
        #one['Date2']=str(one['Date'])
        two=one.groupby(["Month","Year","Day"],as_index=False).min()
        two["Date"]=pd.DatetimeIndex(two['Year'].astype(str)+"-"+two['Month'].astype(str)+"-"+two["Day"].astype(str))
        two=two[["Date","FATES_LEAF_H2OPOT_SZPF_12"]]
        two.columns=["Date","LWP_min"]
        three=one.groupby(["Month","Year","Day"],as_index=False).max()
        three["Date"]=pd.DatetimeIndex(three['Year'].astype(str)+"-"+three['Month'].astype(str)+"-"+three["Day"].astype(str))
        three=three[["Date","FATES_LEAF_H2OPOT_SZPF_12"]]
        three.columns=["Date","LWP_max"]
        two=two.merge(three,on="Date")
        two["Step"]=list(range(0,len(two)))
        two["Model"]=i
        df=pd.concat([df,two])
```
