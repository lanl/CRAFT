```python
import pandas as pd 
import numpy as np
```

```python
Obsdir="C:/Users/345578/Downloads/Compare_Datas/Compare_Datas/"
LWP=pd.read_csv(Obsdir+"SLZ_LWP.csv")
```

```python
LWP["Date"]=pd.to_datetime(LWP["Date"]).dt.date
LWP["Mpa"]=LWP["LWP_bar"]*.1

LWP2=LWP[["Date","Mpa"]]
LWPmin=LWP2.groupby("Date",as_index=False).min()
LWPmin["Error"]=0.1
LWPmin["Set"]="LWP_min"
LWPmin["obs"]=LWPmin["Mpa"]
LWPmax=LWP2.groupby("Date",as_index=False).max()
LWPmax["Error"]=0.1
LWPmax["Set"]="LWP_max"
LWPmax["obs"]=LWPmax["Mpa"]
```

```python
SWC=pd.read_csv(Obsdir+"SLZ_VWC.csv")
SWC["Date"]=pd.to_datetime(SWC["TIMESTAMP"]).dt.date
SWC=SWC[SWC['VWC_10_Avg']!="NAN"]
SWC=SWC[["Date",'VWC_10_Avg']]
#SWC.groupby("Date",as_index=False).mean() 
SWC['VWC_10_Avg']=SWC['VWC_10_Avg'].astype("float")
SWC.groupby("Date",as_index=False).mean() 
SWC["Error"]=0.1
SWC["Set"]="H2OSOI1"
SWC["obs"]='VWC_10_Avg'

SF=pd.read_csv(Obsdir+"SLZ_sapflowrange.csv")
SF=SF[["Date","mean"]]
SF["Error"]=0.1
SF["Set"]="SAPFLOW"
SF['obs']=SF["mean"]


```

```python
Full=pd.concat(
    [SF[["Date","obs","Error","Set"]],
     SWC[["Date","obs","Error","Set"]],
     LWPmax[["Date","obs","Error","Set"]],
     LWPmin[["Date","obs","Error","Set"]]],axis=0)
```
